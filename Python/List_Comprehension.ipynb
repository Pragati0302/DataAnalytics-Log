# List Comprehension

```python
# List comprehension returns a list
numbers = [1, 2, 3, 4]
decremented = [x - 1 for x in numbers]
print("List comprehension:", decremented)

print("Type:", type(decremented))
```

### Generator Expression

```python
# Generator expression returns a generator object
gen_expr = (x - 1 for x in numbers)
print("Generator expression:", gen_expr)

print("Type:", type(gen_expr))

# Convert generator to list
print("As list:", list(gen_expr))
```

### Generator -> List Conversion

```python
# Convert generator expression to tuple
tup = tuple(x - 1 for x in numbers)
print("Tuple from generator:", tup)

print("Type:", type(tup))
```

Q: You are given a tuple numbers that contains integers. You need to return a tuple containing doubles of given numbers.

```python
#Using lists and tuple menthod
def doubleTup(numbers):
    li = []
    for i in numbers:
        li.append(i * 2)
    return tuple(li)

#Time complexity: O(n)
#Space Complexity: O(n)*2  as it creates an intermediate list (li) before converting it to a tuple
```

```python
#Using generators
def doubleTup(numbers):
    return tuple(x * 2 for x in numbers)

#Time complexity: O(n)
#Space Complexity: O(n)
```

- Generator expressions are often faster and more memory-friendly for simple transformations.
- They’re also more Pythonic and concise—great for readability and showcasing clean code in your GitHub projects.

**NOTE: Use generators when working with large datasets to save memory. Use lists or tuples when you need indexed access or immutability.**
